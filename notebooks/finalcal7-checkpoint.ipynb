```python
import os
import cv2
import numpy as np
from skimage.feature import local_binary_pattern
from sklearn.decomposition import PCA

# LBP parameters
radius = 3  # Radius for LBP
n_points = 8 * radius  # Number of points considered for LBP

# Function to extract LBP features from a single image
def extract_lbp_features(image, radius=3, n_points=24):
    # Compute LBP
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    # Compute the LBP histogram
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return lbp_hist

# Function to load images and extract LBP features for each image
def load_images_and_extract_lbp(data_dir):
    lbp_features = []
    image_paths = []

    # Iterate through each file in the dataset directory
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img_path = os.path.join(data_dir, file_name)
            # Load image in grayscale mode
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Store the image path
                image_paths.append(img_path)
                # Extract LBP features
                lbp_hist = extract_lbp_features(img, radius, n_points)
                lbp_features.append(lbp_hist)

    # Convert the list of features to a NumPy array
    lbp_features = np.array(lbp_features)

    return lbp_features, image_paths

# Path to your dataset
data_dir = r"C:\Users\Asus\Desktop\COIL dataset\archive (1)\CAL7"

# Extract LBP features from the dataset
lbp_features, image_paths = load_images_and_extract_lbp(data_dir)

# Display information
print(f"Extracted LBP features for {len(image_paths)} images.")
print(f"Feature shape: {lbp_features.shape}")
```

```python
import os
import cv2
import numpy as np
import pywt

# Function to extract Wavelet Moments from a single image
def extract_wm_features(image):
    # Perform 2D Discrete Wavelet Transform (DWT) on the image
    coeffs2 = pywt.dwt2(image, 'haar')  # Using 'haar' wavelet
    cA, (cH, cV, cD) = coeffs2
    
    # Compute the mean and standard deviation of the approximation and detail coefficients
    features = []
    for coeff in [cA, cH, cV, cD]:
        features.append(np.mean(coeff))
        features.append(np.std(coeff))
    
    return np.array(features)

# Function to load images and extract Wavelet Moment (WM) features for each image
def load_images_and_extract_wm(data_dir):
    wm_features = []
    image_paths = []

    # Iterate through each file in the dataset directory
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img_path = os.path.join(data_dir, file_name)
            # Load image in grayscale mode
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Store the image path
                image_paths.append(img_path)
                # Extract WM features
                wm_features.append(extract_wm_features(img))

    # Convert the list of features to a NumPy array
    wm_features = np.array(wm_features)

    return wm_features, image_paths

# Path to your dataset
data_dir = r"C:\Users\Asus\Desktop\COIL dataset\archive (1)\CAL7"

# Extract WM features from the dataset
wm_features, image_paths = load_images_and_extract_wm(data_dir)

# Display information
print(f"Extracted WM features for {len(image_paths)} images.")
print(f"Feature shape: {wm_features.shape}")
```

```python
import os
import cv2
import numpy as np
from skimage.feature import local_binary_pattern

# CENTRIST uses LBP with 'riu2' method, which computes the census transform
# Function to extract CENTRIST (LBP-based) features from a single image
def extract_centrist_features(image, radius=3, n_points=24):
    # Perform the LBP computation with 'uniform' method
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    # Calculate the histogram of LBP values (which are effectively CENTRIST features)
    centrist_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return centrist_hist

# Function to load images and extract CENTRIST features for each image
def load_images_and_extract_centrist(data_dir):
    centrist_features = []
    image_paths = []

    # Iterate through each file in the dataset directory
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img_path = os.path.join(data_dir, file_name)
            # Load image in grayscale mode
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Store the image path
                image_paths.append(img_path)
                # Extract CENTRIST features
                centrist_hist = extract_centrist_features(img)
                centrist_features.append(centrist_hist)

    # Convert the list of features to a NumPy array
    centrist_features = np.array(centrist_features)

    return centrist_features, image_paths

# Path to your dataset
data_dir = r"C:\Users\Asus\Desktop\COIL dataset\archive (1)\CAL7"

# Extract CENTRIST features from the dataset
centrist_features, image_paths = load_images_and_extract_centrist(data_dir)

# Display information
print(f"Extracted CENTRIST features for {len(image_paths)} images.")
print(f"Feature shape: {centrist_features.shape}")
```

```python
# DBSCAN for LBP Features

from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
import numpy as np

# Define a range of eps values to try
eps_values = np.arange(0.1, 2.1, 0.1)  # Adjust the range as needed
min_samples = 20  # Minimum number of samples in a neighborhood

# Store results
results = []

# Iterate over each eps value
for eps in eps_values:
    # Apply DBSCAN
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(lbp_features_pca)

    # Print the number of clusters found and the number of outliers
    num_clusters = len(set(labels)) - (1 if -1 in labels else 0)  # Exclude noise if present
    num_outliers = list(labels).count(-1)

    print(f"EPS: {eps:.1f} - Number of clusters: {num_clusters}, Number of outliers: {num_outliers}")

    # Calculate the Silhouette Score
    if num_clusters > 1:
        silhouette_avg = silhouette_score(lbp_features_pca, labels)
        print(f"Silhouette Score: {silhouette_avg:.3f}")
        results.append((eps, num_clusters, silhouette_avg))
    else:
        print("Silhouette Score cannot be computed because there is only one cluster.")
        results.append((eps, num_clusters, None))

# Find the optimal eps with the highest silhouette score
optimal_eps = max(results, key=lambda x: x[2] if x[2] is not None else -1)
print(f"Optimal EPS: {optimal_eps[0]:.1f}, Clusters: {optimal_eps[1]}, Silhouette Score: {optimal_eps[2]:.3f}")

# Optionally, plot the silhouette scores for visualization
plt.figure(figsize=(10, 6))
eps_list, silhouette_scores = zip(*[(r[0], r[2]) for r in results if r[2] is not None])
plt.plot(eps_list, silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different EPS Values')
plt.xlabel('EPS')
plt.ylabel('Silhouette Score')
plt.grid()
plt.show()
```

```python
import os
import cv2
import numpy as np
from skimage.feature import local_binary_pattern
import pywt

# LBP parameters
radius = 3  # Radius for LBP
n_points = 8 * radius  # Number of points considered for LBP

# Function to extract LBP features from a single image
def extract_lbp_features(image, radius=3, n_points=24):
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return lbp_hist

# Function to extract Wavelet Moments from a single image
def extract_wm_features(image):
    coeffs2 = pywt.dwt2(image, 'haar')
    cA, (cH, cV, cD) = coeffs2
    features = []
    for coeff in [cA, cH, cV, cD]:
        features.append(np.mean(coeff))
        features.append(np.std(coeff))
    return np.array(features)

# Function to load images, extract LBP and WM features, and concatenate them
def load_images_and_extract_combined_features(data_dir):
    combined_features = []
    image_paths = []

    # Iterate through each file in the dataset directory
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                image_paths.append(img_path)
                lbp_hist = extract_lbp_features(img, radius, n_points)
                wm_features = extract_wm_features(img)
                # Concatenate LBP and WM features
                combined_feature = np.concatenate((lbp_hist, wm_features))
                combined_features.append(combined_feature)

    combined_features = np.array(combined_features)
    return combined_features, image_paths

# Path to your dataset
data_dir = r"C:\Users\Asus\Desktop\COIL dataset\archive (1)\CAL7"

# Extract combined LBP and WM features
combined_features, image_paths = load_images_and_extract_combined_features(data_dir)

# Display information
print(f"Extracted combined features for {len(image_paths)} images.")
print(f"Combined feature shape: {combined_features.shape}")
```

```python
import os
import cv2
import numpy as np
from skimage.feature import local_binary_pattern
import pywt  # PyWavelets for WM feature extraction

# Function to extract CENTRIST (LBP-based) features from a single image
def extract_centrist_features(image, radius=3, n_points=24):
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    centrist_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return centrist_hist

# Function to extract Wavelet Moments (WM) features from a single image
def extract_wm_features(image):
    # Perform 2D Discrete Wavelet Transform (DWT)
    coeffs = pywt.dwt2(image, 'haar')  # Using Haar wavelet
    cA, (cH, cV, cD) = coeffs  # Approximation, Horizontal, Vertical, Diagonal coefficients
    
    # Compute WM as mean and standard deviation of wavelet coefficients
    wm_features = np.hstack([
        np.mean(cA), np.std(cA),
        np.mean(cH), np.std(cH),
        np.mean(cV), np.std(cV),
        np.mean(cD), np.std(cD)
    ])
    return wm_features

# Function to load images and extract combined CENTRIST and WM features
def load_images_and_extract_combined_features(data_dir):
    combined_features = []
    image_paths = []

    # Iterate through each file in the dataset directory
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img_path = os.path.join(data_dir, file_name)
            # Load image in grayscale mode
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Store the image path
                image_paths.append(img_path)
                # Extract CENTRIST features
                centrist_hist = extract_centrist_features(img)
                # Extract WM features
                wm_features = extract_wm_features(img)
                # Combine CENTRIST and WM features
                combined_feature = np.hstack([centrist_hist, wm_features])
                combined_features.append(combined_feature)

    # Convert the list of features to a NumPy array
    combined_features = np.array(combined_features)

    return combined_features, image_paths

# Path to your dataset
data_dir = r"C:\Users\Asus\Desktop\COIL dataset\archive (1)\CAL7"

# Extract combined features from the dataset
combined_features, image_paths = load_images_and_extract_combined_features(data_dir)

# Display information
print(f"Extracted combined features for {len(image_paths)} images.")
print(f"Combined feature shape: {combined_features.shape}")
```

```python
import os
import cv2
import numpy as np
from skimage.feature import local_binary_pattern
import pywt

# LBP parameters
radius = 3  # Radius for LBP
n_points = 8 * radius  # Number of points considered for LBP

# Function to extract LBP features from a single image
def extract_lbp_features(image, radius=3, n_points=24):
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return lbp_hist

# Function to extract Wavelet Moments from a single image
def extract_wm_features(image):
    coeffs2 = pywt.dwt2(image, 'haar')  # Using 'haar' wavelet
    cA, (cH, cV, cD) = coeffs2
    features = []
    for coeff in [cA, cH, cV, cD]:
        features.append(np.mean(coeff))
        features.append(np.std(coeff))
    return np.array(features)

# Function to extract CENTRIST features from a single image
def extract_centrist_features(image, radius=3, n_points=24):
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    centrist_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return centrist_hist

# Function to load images and extract combined features for each image
def load_images_and_extract_combined(data_dir):
    combined_features = []
    image_paths = []

    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                image_paths.append(img_path)
                # Extract features
                lbp_hist = extract_lbp_features(img, radius, n_points)
                wm_features = extract_wm_features(img)
                centrist_hist = extract_centrist_features(img, radius, n_points)

                # Combine all features into a single feature vector
                combined_feature = np.concatenate((lbp_hist, wm_features, centrist_hist))
                combined_features.append(combined_feature)

    # Convert the list of combined features to a NumPy array
    combined_features = np.array(combined_features)

    return combined_features, image_paths

# Path to your dataset
data_dir = r"C:\Users\Asus\Desktop\COIL dataset\archive (1)\CAL7"

# Extract combined features from the dataset
combined_features, image_paths = load_images_and_extract_combined(data_dir)

# Display information
print(f"Extracted combined features for {len(image_paths)} images.")
print(f"Feature shape: {combined_features.shape}")
```

```python
import os
import cv2
import numpy as np
from skimage.feature import local_binary_pattern
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pywt

# LBP parameters
radius = 3  # Radius for LBP
n_points = 8 * radius  # Number of points considered for LBP

# Function to extract LBP features from a single image
def extract_lbp_features(image, radius=3, n_points=24):
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return lbp_hist

# Function to extract Wavelet Moments from a single image
def extract_wm_features(image):
    coeffs2 = pywt.dwt2(image, 'haar')  # Using 'haar' wavelet
    cA, (cH, cV, cD) = coeffs2
    features = []
    for coeff in [cA, cH, cV, cD]:
        features.append(np.mean(coeff))
        features.append(np.std(coeff))
    return np.array(features)

# Function to extract CENTRIST features from a single image
def extract_centrist_features(image, radius=3, n_points=24):
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    centrist_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return centrist_hist

# Function to load images and extract combined features from LBP, WM, and CENTRIST
def load_images_and_extract_combined_features(data_dir):
    combined_features = []
    image_paths = []

    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                image_paths.append(img_path)
                
                # Extract features
                lbp_hist = extract_lbp_features(img, radius, n_points)
                wm_features = extract_wm_features(img)
                centrist_hist = extract_centrist_features(img, radius, n_points)
                
                # Combine features
                combined_feature = np.concatenate((lbp_hist, wm_features, centrist_hist))
                combined_features.append(combined_feature)

    combined_features = np.array(combined_features)
    return combined_features, image_paths

# Path to your dataset
data_dir = r"C:\Users\Asus\Desktop\COIL dataset\archive (1)\CAL7"

# Extract combined features from LBP, WM, and CENTRIST
combined_features, image_paths = load_images_and_extract_combined_features(data_dir)

# Step 1: Standardize the combined features
scaler = StandardScaler()
combined_features_scaled = scaler.fit_transform(combined_features)

# Step 2: Apply PCA to reduce dimensionality to 2 components (2D)
pca = PCA(n_components=2)
combined_features_pca = pca.fit_transform(combined_features_scaled)

# Display the shape of the PCA-reduced combined features
print(f"Combined LBP, WM, and CENTRIST features after PCA reduction shape: {combined_features_pca.shape}")
```

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
import numpy as np

# Assuming 'combined_features_pca' contains the PCA-reduced features from LBP, CENTRIST, and WM
# Define a range of eps values to try
eps_values = np.arange(0.1, 2.1, 0.1)  # Adjust the range as needed
min_samples = 20  # Minimum number of samples in a neighborhood

# Store results
results = []

# Iterate over each eps value for combined LBP, CENTRIST, and WM features
for eps in eps_values:
    # Apply DBSCAN
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(combined_features_pca)

    # Print the number of clusters found and the number of outliers
    num_clusters = len(set(labels)) - (1 if -1 in labels else 0)  # Exclude noise if present
    num_outliers = list(labels).count(-1)

    print(f"EPS: {eps:.1f} - Number of clusters: {num_clusters}, Number of outliers: {num_outliers}")

    # Calculate the Silhouette Score
    if num_clusters > 1:
        silhouette_avg = silhouette_score(combined_features_pca, labels)
        print(f"Silhouette Score: {silhouette_avg:.3f}")
        results.append((eps, num_clusters, silhouette_avg))
    else:
        print("Silhouette Score cannot be computed because there is only one cluster.")
        results.append((eps, num_clusters, None))

# Find the optimal eps with the highest silhouette score
optimal_eps = max(results, key=lambda x: x[2] if x[2] is not None else -1)
print(f"Optimal EPS: {optimal_eps[0]:.1f}, Clusters: {optimal_eps[1]}, Silhouette Score: {optimal_eps[2]:.3f}")

# Optionally, plot the silhouette scores for visualization
plt.figure(figsize=(10, 6))
eps_list, silhouette_scores = zip(*[(r[0], r[2]) for r in results if r[2] is not None])
plt.plot(eps_list, silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different EPS Values (Combined LBP, CENTRIST, and WM Features)')
plt.xlabel('EPS')
plt.ylabel('Silhouette Score')
plt.grid()
plt.show()
```
