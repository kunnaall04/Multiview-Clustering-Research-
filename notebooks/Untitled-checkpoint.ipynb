```python
import os
import cv2
import numpy as np
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern

# LBP parameters
radius = 3
n_points = 8 * radius

# Function to extract LBP features from an image
def extract_lbp_features(image, radius=3, n_points=24):
    lbp = local_binary_pattern(image, n_points, radius, method='uniform')
    # Compute the histogram of LBP features
    lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
    return lbp_hist

# Function to load images, extract LBP features, and reduce dimensions if necessary
def load_images_and_extract_features(data_dir, max_components=20):
    lbp_features = []
    image_paths = []
    
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png') or file_name.endswith('.jpg'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                image_paths.append(img_path)
                lbp_hist = extract_lbp_features(img, radius, n_points)
                lbp_features.append(lbp_hist)
    
    lbp_features = np.array(lbp_features)
    
    # Reduce dimensions using PCA if the number of features is too high
    if lbp_features.shape[1] > max_components:
        print(f"Reducing dimensions from {lbp_features.shape[1]} to {max_components} using PCA.")
        pca = PCA(n_components=max_components)
        lbp_features = pca.fit_transform(lbp_features)
    
    return lbp_features, image_paths

# Function to apply DBSCAN and plot the clusters
def apply_dbscan_and_plot(features, eps=0.5, min_samples=5):
    # Apply DBSCAN
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(features)

    # Plot the clusters
    plt.figure(figsize=(10, 7))
    unique_labels = set(labels)
    
    for label in unique_labels:
        cluster_points = features[labels == label]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f"Cluster {label}" if label != -1 else "Noise")
    
    plt.title("DBSCAN Clustering")
    plt.xlabel("Feature 1 (or PCA 1)")
    plt.ylabel("Feature 2 (or PCA 2)")
    plt.legend()
    plt.show()

# Main execution
data_dir = r"C:\Users\Asus\Desktop\COIL\archive (1)\CAL7"  # Dataset path
max_components = 20  # Set the desired number of PCA components for dimensionality reduction

# Load images, extract LBP features, and apply PCA if needed
lbp_features, image_paths = load_images_and_extract_features(data_dir, max_components=max_components)

# Apply DBSCAN and plot clusters
apply_dbscan_and_plot(lbp_features, eps= 0.03392037895722123, min_samples=40)

```
