```python
import cv2
import os
import numpy as np
from skimage.feature import local_binary_pattern

def load_and_extract_lbp_features(data_dir, radius=3, n_points=24):
    lbp_features = []
    images = []
    
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Ensure the image is of type uint8
                img = img.astype(np.uint8)
                
                # Extract LBP features
                lbp = local_binary_pattern(img, n_points, radius, method='uniform')
                # Compute histogram of LBP features
                lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
                lbp_features.append(lbp_hist)
                images.append(img)
    
    return np.array(lbp_features), images

# Path to the processed images folder
data_dir =r"C:\Users\Asus\Desktop\COIL\coil-20\coil20_cars_"

# Load images and extract LBP features
lbp_features, images = load_and_extract_lbp_features(data_dir)

# Verify the shape of LBP features
print(f"Number of images processed: {len(images)}")
print(f"Shape of LBP features: {lbp_features.shape}")
```

```python
from sklearn.decomposition import PCA

# Initialize PCA for reducing to 2D
pca = PCA(n_components=2)

# Fit PCA and transform the data
lbp_features_pca = pca.fit_transform(lbp_features)

# Check the shape of PCA result
print(f"Shape of PCA features: {lbp_features_pca.shape}")
```

```python
import matplotlib.pyplot as plt

# Plot PCA results in 2D
plt.figure(figsize=(10, 7))
plt.scatter(lbp_features_pca[:, 0], lbp_features_pca[:, 1], c='blue', marker='o')
plt.title('PCA of LBP Features (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load images and extract LBP features
def load_and_extract_lbp_features(data_dir, radius=3, n_points=24):
    lbp_features = []
    images = []
    file_paths = []  # To store full file paths
    
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Ensure the image is of type uint8
                img = img.astype(np.uint8)
                
                # Extract LBP features
                lbp = local_binary_pattern(img, n_points, radius, method='uniform')
                # Compute histogram of LBP features
                lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
                lbp_features.append(lbp_hist)
                images.append(img)
                file_paths.append(img_path)  # Store the full file path
    
    return np.array(lbp_features), images, file_paths

# Path to the processed images folder
data_dir = r"C:\Users\Asus\Desktop\COIL\coil-20\coil20_cars_"

# Load images and extract LBP features
lbp_features, images, file_paths = load_and_extract_lbp_features(data_dir)

# Verify the shape of LBP features
print(f"Number of images processed: {len(images)}")
print(f"Shape of LBP features: {lbp_features.shape}")

# Initialize PCA for reducing to 2D
pca = PCA(n_components=2)

# Fit PCA and transform the data
lbp_features_pca = pca.fit_transform(lbp_features)

# Check the shape of PCA result
print(f"Shape of PCA features: {lbp_features_pca.shape}")

# Plot PCA results in 2D
plt.figure(figsize=(10, 7))
plt.scatter(lbp_features_pca[:, 0], lbp_features_pca[:, 1], c='blue', marker='o')
plt.title('PCA of LBP Features (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Set environment variable to avoid memory leak warning
os.environ["OMP_NUM_THREADS"] = "6"

# Number of clusters
n_clusters = 2

# Initialize KMeans with the number of clusters and set n_init explicitly
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)

# Fit KMeans to the PCA-transformed features
clusters = kmeans.fit_predict(lbp_features_pca)

# Add cluster centers
cluster_centers = kmeans.cluster_centers_

print(f"Cluster centers shape: {cluster_centers.shape}")
print(f"Number of clusters: {n_clusters}")

# Plot PCA results with K-Means clusters
plt.figure(figsize=(10, 7))
plt.scatter(lbp_features_pca[:, 0], lbp_features_pca[:, 1], c=clusters, cmap='viridis', marker='o')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.title('PCA of LBP Features with K-Means Clustering (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()

# Extract file paths based on cluster assignments
clustered_file_paths = {i: [] for i in range(n_clusters)}
for file_path, cluster in zip(file_paths, clusters):
    clustered_file_paths[cluster].append(file_path)

# Print the total number of data points and the number of points in each cluster
print(f"Total number of data points: {len(file_paths)}")
for cluster, files in clustered_file_paths.items():
    print(f"Number of points in Cluster {cluster + 1}: {len(files)}")
    print(f"File paths in Cluster {cluster + 1}:")
    for file in files:
        print(file)
```
