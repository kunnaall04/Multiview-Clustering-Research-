```python
import os
import cv2
import numpy as np
from skimage.feature import local_binary_pattern, hog
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# LBP Feature Extraction Function
def extract_lbp_features(image, P=8, R=1, method='uniform'):
    lbp = local_binary_pattern(image, P, R, method)
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, P + 3), range=(0, P + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-6)
    return hist

# HOG Feature Extraction Function
def extract_hog_features(image, pixels_per_cell=(8, 8), cells_per_block=(2, 2), orientations=9):
    hog_features, hog_image = hog(image, pixels_per_cell=pixels_per_cell, 
                                  cells_per_block=cells_per_block, orientations=orientations, 
                                  visualize=True, block_norm='L2-Hys')
    return hog_features

# Load Images and Extract Combined LBP and HOG Features
def load_and_extract_features(data_dir):
    lbp_features_list = []
    hog_features_list = []
    file_paths = []
    
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                img = img.astype(np.uint8)
                
                # Extract LBP features
                lbp_features = extract_lbp_features(img)
                lbp_features_list.append(lbp_features)
                
                # Extract HOG features
                hog_features = extract_hog_features(img)
                hog_features_list.append(hog_features)
                
                file_paths.append(img_path)  # Store the full file path
    
    lbp_features = np.array(lbp_features_list)
    hog_features = np.array(hog_features_list)
    
    # Combine LBP and HOG features
    combined_features = np.hstack((lbp_features, hog_features))
    
    return combined_features, file_paths

# Path to the dataset
data_dir = r"C:\Users\Asus\Desktop\COIL\coil-20\coil20_cars_"

# Load images and extract combined features
combined_features, file_paths = load_and_extract_features(data_dir)

# Verify the shape of combined features
print(f"Shape of combined features: {combined_features.shape}")

# PCA for Dimensionality Reduction to 2D
pca = PCA(n_components=2)
combined_features_pca = pca.fit_transform(combined_features)
print(f"Shape of PCA features: {combined_features_pca.shape}")

# KMeans Clustering with 2 Clusters
n_clusters = 2
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
clusters = kmeans.fit_predict(combined_features_pca)
cluster_centers = kmeans.cluster_centers_

print(f"Cluster centers shape: {cluster_centers.shape}")
print(f"Cluster assignments: {clusters}")

# Calculate silhouette score
silhouette_avg = silhouette_score(combined_features_pca, clusters)
print(f"Silhouette Score: {silhouette_avg:.4f}")

# Plot PCA results in 2D with centroids
plt.figure(figsize=(10, 7))
plt.scatter(combined_features_pca[:, 0], combined_features_pca[:, 1], c=clusters, cmap='viridis', marker='o')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=300, c='red', marker='X')  # Plot centroids
plt.title('PCA of Combined LBP and HOG Features (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.show()

# Group files by cluster
clustered_files = {i: [] for i in range(n_clusters)}
for idx, cluster_id in enumerate(clusters):
    clustered_files[cluster_id].append(file_paths[idx])

# Print the full file paths in each cluster
for cluster_id, files in clustered_files.items():
    print(f"Number of points in Cluster {cluster_id + 1}: {len(files)}")
    print(f"Full file paths in Cluster {cluster_id + 1}:")
    for file in files:
        print(file)
    print("\n")
```
