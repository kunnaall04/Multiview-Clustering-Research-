```python
import cv2
import os
import numpy as np

def load_coil20_images(data_dir):
    images = []
    labels = []
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                images.append(img)
                try:
                    # Extract label from file name, assuming format like "obj01_01.png"
                    label = int(file_name.split('_')[0][3:])
                    labels.append(label)
                except ValueError:
                    print(f"Could not extract label from file: {file_name}")
            else:
                print(f"Failed to load image: {file_name}")
        else:
            print(f"File is not a .png image: {file_name}")
    return images, labels

# Path to the processed images folder
data_dir = r"C:\Users\Asus\Desktop\COIL\coil-20\coil-20-proc"

# Load images and labels
images, labels = load_coil20_images(data_dir)

# Verify the length of the lists and the first item
print(f"Number of images loaded: {len(images)}")
if len(images) > 0:
    print(f"First image shape: {images[0].shape}")
    print(f"First label: {labels[0]}")
else:
    print("No images loaded.")
```

```python
import cv2
import os
import numpy as np
from skimage.feature import local_binary_pattern

def load_and_extract_lbp_features(data_dir, radius=3, n_points=24):
    lbp_features = []
    images = []
    
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Ensure the image is of type uint8
                img = img.astype(np.uint8)
                
                # Extract LBP features
                lbp = local_binary_pattern(img, n_points, radius, method='uniform')
                # Compute histogram of LBP features
                lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
                lbp_features.append(lbp_hist)
                images.append(img)
    
    return np.array(lbp_features), images

# Path to the processed images folder
data_dir = r"C:\Users\Asus\Desktop\COIL\coil-20\coil-20-proc"

# Load images and extract LBP features
lbp_features, images = load_and_extract_lbp_features(data_dir)

# Verify the shape of LBP features
print(f"Number of images processed: {len(images)}")
print(f"Shape of LBP features: {lbp_features.shape}")
```

```python
from sklearn.decomposition import PCA

# Initialize PCA for reducing to 2D
pca = PCA(n_components=2)

# Fit PCA and transform the data
lbp_features_pca = pca.fit_transform(lbp_features)

# Check the shape of PCA result
print(f"Shape of PCA features: {lbp_features_pca.shape}")
```

```python
import matplotlib.pyplot as plt

# Plot PCA results in 2D
plt.figure(figsize=(10, 7))
plt.scatter(lbp_features_pca[:, 0], lbp_features_pca[:, 1], c='blue', marker='o')
plt.title('PCA of LBP Features (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Set environment variable to avoid memory leak warning
import os
os.environ["OMP_NUM_THREADS"] = "6"

# Number of clusters
n_clusters = 5

# Initialize KMeans with the number of clusters and set n_init explicitly
kmeans = KMeans(n_clusters=n_clusters, n_init=10, random_state=42)

# Fit KMeans to the PCA-transformed features
clusters = kmeans.fit_predict(lbp_features_pca)

# Add cluster centers
cluster_centers = kmeans.cluster_centers_

print(f"Cluster centers shape: {cluster_centers.shape}")
print(f"Number of clusters: {n_clusters}")
```

```python
from skimage.feature import hog
from skimage import exposure

def extract_hog_features(image, pixels_per_cell=(8, 8), cells_per_block=(2, 2), orientations=9):
    # Compute HOG features
    hog_features, hog_image = hog(image, pixels_per_cell=pixels_per_cell, 
                                  cells_per_block=cells_per_block, orientations=orientations, 
                                  visualize=True, block_norm='L2-Hys')
    # Normalize the HOG image for visualization
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    return hog_features, hog_image_rescaled

def load_and_extract_hog_features(data_dir):
    hog_features_list = []
    images = []
    
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                img = img.astype(np.uint8)
                # Extract HOG features
                hog_features, _ = extract_hog_features(img)
                hog_features_list.append(hog_features)
                images.append(img)
    
    return np.array(hog_features_list), images

# Path to the processed images folder
data_dir = r"C:\Users\Asus\Desktop\COIL\coil-20\coil-20-proc"

# Load images and extract HOG features
hog_features, images = load_and_extract_hog_features(data_dir)

# Verify the shape of HOG features
print(f"Number of images processed: {len(images)}")
print(f"Shape of HOG features: {hog_features.shape}")
```

```python
from sklearn.decomposition import PCA

# Initialize PCA for reducing to 2D
pca = PCA(n_components=2)

# Fit PCA and transform the data
hog_features_pca = pca.fit_transform(hog_features)

# Check the shape of PCA result
print(f"Shape of PCA features: {hog_features_pca.shape}")
```

```python
from sklearn.cluster import KMeans

# Initialize KMeans with the number of clusters
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit KMeans to the PCA-transformed features
clusters = kmeans.fit_predict(hog_features_pca)

# Add cluster centers
cluster_centers = kmeans.cluster_centers_

print(f"Cluster centers shape: {cluster_centers.shape}")
print(f"Cluster assignments: {clusters}")
```

```python
import matplotlib.pyplot as plt

# Plot PCA results in 2D
plt.figure(figsize=(10, 7))
plt.scatter(hog_features_pca[:, 0], hog_features_pca[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title('PCA of HOG Features (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.show()
```
