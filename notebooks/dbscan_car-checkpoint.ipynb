```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern

def load_coil20_images(data_dir):
    images = []
    labels = []
    
    # Sorting the filenames to maintain a consistent order
    for file_name in sorted(os.listdir(data_dir)):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                images.append(img)
                
                # Extract label from the filename, assuming format like "obj1_0degree.png"
                try:
                    label = int(file_name.split('_')[0][3:])
                    labels.append(label)
                except ValueError:
                    print(f"Could not extract label from file: {file_name}")
            else:
                print(f"Failed to load image: {file_name}")
        else:
            print(f"File is not a .png image: {file_name}")
            
    return images, labels

# **Update the path here to the correct location**
data_dir = r"C:\Users\Asus\Desktop\COIL\coil-20\coil20_cars_"

# Load images and labels
images, labels = load_coil20_images(data_dir)

# Verify the length of the lists and the first item
print(f"Number of images loaded: {len(images)}")
if len(images) > 0:
    print(f"First image shape: {images[0].shape}")
    print(f"First label: {labels[0]}")
else:
    print("No images loaded.")

# Parameters for LBP
radius = 3  # LBP radius
n_points = 8 * radius  # Number of points for LBP

# Extract LBP for a single image (first image in the list)
image = images[10]
lbp = local_binary_pattern(image, n_points, radius, method='uniform')

# Plot the original image and its LBP
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(lbp, cmap='gray')
plt.title('LBP Image')
plt.show()

# Plot the LBP histogram
plt.hist(lbp.ravel(), bins=range(0, n_points + 3), density=True)
plt.title('LBP Histogram')
plt.show()
```

```python
import cv2
import os
import numpy as np
from skimage.feature import local_binary_pattern
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

def load_and_extract_lbp_features(data_dir, radius=3, n_points=24):
    lbp_features = []
    images = []
    
    for file_name in os.listdir(data_dir):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Ensure the image is of type uint8
                img = img.astype(np.uint8)
                
                # Extract LBP features
                lbp = local_binary_pattern(img, n_points, radius, method='uniform')
                # Compute histogram of LBP features
                lbp_hist, _ = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), density=True)
                lbp_features.append(lbp_hist)
                images.append(img)
    
    return np.array(lbp_features), images

# Path to the processed images folder
data_dir = r"C:\Users\Asus\Desktop\COIL\coil-20\coil20_cars_"
# Load images and extract LBP features
lbp_features, images = load_and_extract_lbp_features(data_dir)

# Verify the shape of LBP features
print(f"Number of images processed: {len(images)}")
print(f"Shape of LBP features: {lbp_features.shape}")

# Initialize PCA for reducing to 2D
pca = PCA(n_components=2)

# Fit PCA and transform the data
lbp_features_pca = pca.fit_transform(lbp_features)

# Check the shape of PCA result
print(f"Shape of PCA features: {lbp_features_pca.shape}")

# Apply DBSCAN for clustering
eps =0.05648455162678052 # Maximum distance for points to be considered neighbors
min_samples = 20  # Minimum number of points to form a cluster
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan.fit_predict(lbp_features_pca)

# Check the unique labels assigned by DBSCAN
print(f"Unique labels assigned by DBSCAN: {np.unique(dbscan_labels)}")

# Plot PCA results and color by DBSCAN cluster labels
plt.figure(figsize=(10, 7))
scatter = plt.scatter(lbp_features_pca[:, 0], lbp_features_pca[:, 1], c=dbscan_labels, cmap='viridis', marker='o')
plt.title('DBSCAN Clustering of LBP Features (PCA 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt

# Nearest Neighbors plot
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(lbp_features_pca)
distances, indices = neighbors_fit.kneighbors(lbp_features_pca)

# Sort distances and plot them
distances = np.sort(distances[:, 4])
plt.plot(distances)
plt.title('K-nearest Neighbors Distance for Eps Selection')
plt.show()
```

```python
# Calculate the difference between successive distances
differences = np.diff(distances)

# Find the index of the maximum difference (elbow point)
elbow_index = np.argmax(differences)

# Suggested eps value (distance at the elbow)
suggested_eps = distances[elbow_index]

print(f"Suggested eps value: {suggested_eps}")
```

```python
import cv2
import os
import numpy as np
from skimage.feature import hog
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

def load_and_extract_hog_features(data_dir):
    hog_features = []
    images = []
    
    # Parameters for HOG
    hog_params = {
        'orientations': 9, 
        'pixels_per_cell': (8, 8), 
        'cells_per_block': (2, 2), 
        'block_norm': 'L2-Hys', 
        'transform_sqrt': True
    }
    
    for file_name in sorted(os.listdir(data_dir)):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Compute HOG features
                hog_feature = hog(img, **hog_params)
                hog_features.append(hog_feature)
                images.append(img)
    
    return np.array(hog_features), images

# Path to the processed images folder
data_dir =r"C:\Users\Asus\Desktop\COIL\coil-20\coil20_cars_"

# Load images and extract HOG features
hog_features, images = load_and_extract_hog_features(data_dir)

# Verify the shape of HOG features
print(f"Number of images processed: {len(images)}")
print(f"Shape of HOG features: {hog_features.shape}")

# Initialize PCA for reducing to 2D
pca = PCA(n_components=2)

# Fit PCA and transform the data
hog_features_pca = pca.fit_transform(hog_features)

# Check the shape of PCA result
print(f"Shape of PCA features: {hog_features_pca.shape}")

# Apply DBSCAN for clustering
eps = 0.5  # Maximum distance for points to be considered neighbors (adjust if needed)
min_samples = 5  # Minimum number of points to form a cluster
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan.fit_predict(hog_features_pca)

# Check the unique labels assigned by DBSCAN
print(f"Unique labels assigned by DBSCAN: {np.unique(dbscan_labels)}")

# Plot PCA results and color by DBSCAN cluster labels
plt.figure(figsize=(10, 7))
scatter = plt.scatter(hog_features_pca[:, 0], hog_features_pca[:, 1], c=dbscan_labels, cmap='viridis', marker='o')
plt.title('DBSCAN Clustering of HOG Features (PCA 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```

```python
import cv2
import numpy as np
from skimage.feature import hog, local_binary_pattern
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import os

# Parameters for LBP
radius = 1
n_points = 8 * radius
method = 'uniform'

# Parameters for HOG
hog_params = {
    'orientations': 9, 
    'pixels_per_cell': (8, 8), 
    'cells_per_block': (2, 2), 
    'block_norm': 'L2-Hys', 
    'transform_sqrt': True
}

def extract_lbp_features(image):
    lbp = local_binary_pattern(image, n_points, radius, method)
    # Create a histogram of LBP features
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + 1e-7)  # Normalize the histogram
    return hist

def extract_hog_features(image):
    return hog(image, **hog_params)

def load_and_extract_features(data_dir):
    combined_features = []
    images = []
    
    for file_name in sorted(os.listdir(data_dir)):
        if file_name.endswith('.png'):
            img_path = os.path.join(data_dir, file_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                # Extract HOG and LBP features
                hog_feature = extract_hog_features(img)
                lbp_feature = extract_lbp_features(img)
                
                # Combine HOG and LBP features
                combined_feature = np.hstack([hog_feature, lbp_feature])
                
                combined_features.append(combined_feature)
                images.append(img)
    
    return np.array(combined_features), images

# Path to the processed images folder
data_dir =r"C:\Users\Asus\Desktop\COIL\coil-20\coil20_cars_"


# Load images and extract HOG + LBP features
combined_features, images = load_and_extract_features(data_dir)

# Verify the shape of combined features
print(f"Number of images processed: {len(images)}")
print(f"Shape of combined HOG + LBP features: {combined_features.shape}")

# Initialize PCA for reducing to 2D
pca = PCA(n_components=2)

# Fit PCA and transform the combined features
combined_features_pca = pca.fit_transform(combined_features)

# Check the shape of PCA result
print(f"Shape of PCA features: {combined_features_pca.shape}")

# Apply DBSCAN for clustering
eps = 0.5  # Adjust eps for DBSCAN if necessary
min_samples = 5  # Minimum samples for DBSCAN
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan_labels = dbscan.fit_predict(combined_features_pca)

# Check the unique labels assigned by DBSCAN
print(f"Unique labels assigned by DBSCAN: {np.unique(dbscan_labels)}")

# Plot PCA results and color by DBSCAN cluster labels
plt.figure(figsize=(10, 7))
scatter = plt.scatter(combined_features_pca[:, 0], combined_features_pca[:, 1], c=dbscan_labels, cmap='viridis', marker='o')
plt.title('DBSCAN Clustering of Combined HOG + LBP Features (PCA 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter, label='Cluster Label')
plt.show()
```
